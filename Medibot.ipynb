```python
import re

def extract_symptoms(text):
    symptoms_db = [
        "fever", "cough", "chest pain", "breathing difficulty",
        "fatigue", "weight loss", "night sweats"
    ]
    
    found = []
    text = text.lower()
    
    for symptom in symptoms_db:
        if symptom in text:
            found.append(symptom)
    
    return found
```

```python
def predict_condition(symptoms):
    if "cough" in symptoms and "weight loss" in symptoms and "night sweats" in symptoms:
        return "Possible Tuberculosis"
    
    if "chest pain" in symptoms and "breathing difficulty" in symptoms:
        return "Possible Lung Infection"
    
    if "fever" in symptoms and "cough" in symptoms:
        return "Possible Respiratory Infection"
    
    return "General Health Issue"
```

```python
def detect_severity(symptoms):
    high_risk = ["breathing difficulty", "severe chest pain"]
    
    for s in symptoms:
        if s in high_risk:
            return "HIGH"
    
    return "LOW"
```

```python
def generate_advice(symptoms):
    advice = []
    
    if "fever" in symptoms:
        advice.append("Take paracetamol if fever is high")
    
    if "cough" in symptoms:
        advice.append("Avoid cold drinks and cover mouth while coughing")
    
    if "weight loss" in symptoms:
        advice.append("Increase nutritional intake and monitor weight")
    
    return advice
```

```python
def generate_response(symptoms):
    condition = predict_condition(symptoms)
    severity = detect_severity(symptoms)
    advice_list = generate_advice(symptoms)
    
    advice_text = "\n".join([f"- {a}" for a in advice_list])
    
    response = f"""
🩺 Possible Condition: {condition}
⚠️ Risk Level: {severity}

💡 Personalized Advice:
{advice_text}

🚨 Warning:
- If symptoms worsen
- If breathing difficulty increases

👉 Visit nearest healthcare center

⚠️ This is not a medical diagnosis.
"""
    
    if severity == "HIGH":
        response += "\n🚨 URGENT: Seek medical help immediately."
    
    return response
```

```python
def check_tb_risk(symptoms):
    tb_keywords = ["cough", "fever", "weight loss", "night sweats"]
    match_count = sum(1 for s in symptoms if s in tb_keywords)
    return match_count >= 2
```
